```python
# Question 1: Create a set of first 10 natural numbers
print("Question 1: Create a set of first 10 natural numbers")

first_natural=set()

for num in range(1,11):
    first_natural.add(num)

print(f"First 10 Natural number is:- {first_natural}")
```

```python
# Question 8: Create a set of vowels from string "hello world"
print("\nQuestion 8: Create a set of vowels from string 'hello world'")

slogan= "hello world"
letter=set()
vowels="aeiou"

for char in slogan.lower():
    if char in vowels:
        letter.add(char)
print(f"The Vowels in {slogan} is:- {letter}")
        
    
```
